```python
gwnolst = list()

# script to generate gwno datamap
# generate gwno country name + gwno code pair list
for gwno in wep['gwno'].unique():
    if pax[pax['Loc1GWNO'] == gwno].shape[0] > 0 or pax[pax['Loc2GWNO'] == gwno].shape[0] > 0:
        country_names = (wep[wep['gwno'] == gwno]['country'])
        if country_names.size > 0:
            gwnolst.append([country_names.iloc[0], gwno])

df = pd.DataFrame(np.array(gwnolst), columns=['country', 'gwno'])
df.to_csv('../../data/gwno.csv', index=False)
df
```
